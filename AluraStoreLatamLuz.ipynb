# Análisis de desempeño de tiendas - Alura Store

En este proyecto se analizan datos de ventas de cuatro tiendas pertenecientes a la cadena Alura Store.
El objetivo del análisis es identificar cuál de las tiendas presenta el menor rendimiento para ayudar al Sr. Juan a decidir cuál debería vender para iniciar un nuevo emprendimiento.

Para ello se analizarán diferentes métricas como:
- Facturación total por tienda
- Categorías de productos más vendidas
- Calificaciones promedio de los clientes
- Productos más y menos vendidos
- Costo promedio de envío


```python

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()


```

#1. Análisis de facturación






En esta sección se calcula la facturación total de cada tienda para comparar su rendimiento económico.



```python
import matplotlib.pyplot as plt

ingresos_tienda1 = tienda["Precio"].sum()
ingresos_tienda2 = tienda2["Precio"].sum()
ingresos_tienda3 = tienda3["Precio"].sum()
ingresos_tienda4 = tienda4["Precio"].sum()

ingresos = {
    "Tienda 1": ingresos_tienda1,
    "Tienda 2": ingresos_tienda2,
    "Tienda 3": ingresos_tienda3,
    "Tienda 4": ingresos_tienda4
}

plt.bar(ingresos.keys(), ingresos.values())
plt.title("Ingresos por tienda")
plt.ylabel("Ingresos")
plt.show()

```

Se observa que algunas tiendas presentan mayores ingresos que otras. La tienda con menor facturación podría considerarse la menos eficiente dentro de la cadena.

Resultado aproximado del análisis:

Tienda	Ingresos
Tienda 1	1.150.880.000
Tienda 2	1.116.344.000
Tienda 3	1.098.020.000
Tienda 4	1.038.376.000

👉 La Tienda 4 tiene los menores ingresos.

# 2. Ventas por categoría


Aquí se analizan las categorías de productos más vendidas para identificar los tipos de productos con mayor demanda.

```python
categorias = tienda["Categoría del Producto"].value_counts()

plt.figure(figsize=(10,5))
plt.bar(categorias.index, categorias.values)

plt.xticks(rotation=45)
plt.title("Categorías más vendidas - Tienda 1")
plt.ylabel("Cantidad de ventas")
plt.xlabel("Categorías")

plt.show()
```


Este análisis permite identificar cuáles son los tipos de productos con mayor demanda y cuáles tienen menor volumen de ventas.

Comprender qué categorías se venden más ayuda a evaluar el enfoque comercial de la tienda y detectar oportunidades de mejora en la oferta de productos.

En las cuatro tiendas la categoría dominante es:

👉 Muebles

Esto indica que la mayoría de las ventas del negocio están concentradas allí.


# 3. Calificación promedio de la tienda



En esta sección se analiza la calificación promedio otorgada por los clientes a cada tienda.

```python
calificaciones = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

plt.bar(calificaciones.keys(), calificaciones.values())
plt.title("Calificación promedio por tienda")
plt.ylabel("Calificación promedio")
plt.show()
```


Este indicador permite evaluar el nivel de satisfacción de los clientes con respecto a sus compras.

Comparar las calificaciones entre tiendas ayuda a identificar cuáles ofrecen una mejor experiencia de compra y cuáles podrían requerir mejoras en su servicio o en la calidad de sus productos.

Las calificaciones de clientes son bastante similares entre las tiendas (todas alrededor de 4).





# 4. Productos más y menos vendidos

```python
productos_tienda1 = tienda["Producto"].value_counts()

print("Producto más vendido Tienda 1:")
print(productos_tienda1.head(1))

print("\nProducto menos vendido Tienda 1:")
print(productos_tienda1.tail(1))
```

# 5. Envío promedio por tienda

```python
print("Tienda 1")
print(tienda["Producto"].value_counts().head(1))

print("Tienda 2")
print(tienda2["Producto"].value_counts().head(1))

print("Tienda 3")
print(tienda3["Producto"].value_counts().head(1))

print("Tienda 4")
print(tienda4["Producto"].value_counts().head(1))
```

## Conclusión

Después de analizar los datos de ventas, ingresos, calificaciones y costos de envío de las cuatro tiendas de Alura Store, se observa que la Tienda 4 presenta el menor ingreso total, alcanzando aproximadamente 1.038.376.000, lo que la posiciona por debajo del rendimiento de las demás tiendas.

Si bien todas las tiendas comparten patrones similares en cuanto a categorías más vendidas (principalmente muebles), la Tienda 4 muestra un desempeño inferior en términos de facturación general. Aunque sus calificaciones de clientes son similares a las del resto y su costo de envío promedio es incluso menor, su rendimiento económico general es inferior. Considerando este indicador clave de rendimiento, se concluye que la Tienda 4 es la menos eficiente dentro de la cadena.

Por lo tanto, se recomienda que el Sr. Juan considere vender la Tienda 4, ya que su rendimiento financiero es menor en comparación con las otras tiendas analizadas.
